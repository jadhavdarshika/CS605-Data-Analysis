```python
import matplotlib.pyplot as plt
```

```python
data = {"Work": 8, "Study": 6, "Leisure": 4, "Sleep": 6}
activities = list(data.keys())
hours = list(data.values())
```

## Bar Chart

```python
plt.bar(activities, hours)
plt.title("Hours Spent on Various Activities")
plt.xlabel("Activities")
plt.ylabel("Hours")
plt.show()
```

## Line Chart

```python
plt.plot(activities, hours, marker="o")
plt.title("Hours Spent on Various Activities")
plt.xlabel("Activities")
plt.ylabel("Hours")
plt.show()
```

## Pie Chart 

```python
plt.pie(hours, labels=activities)
plt.title("Hours Spent on Various Activities")
plt.show()
```
