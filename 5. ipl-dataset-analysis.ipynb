```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# load the IPL dataset
data = pd.read_csv("matches.csv")
```

```python
# Who won most Player of the Match?
player_of_match_count = data['player_of_match'].value_counts()
most_player_of_match = player_of_match_count.index[0]

print("Player with most Player of the Match awards:", most_player_of_match)
```

```python
# Winning Analysis according to Toss
toss_winner_match_winner_count = (data[data['toss_winner'] == data['winner']]['winner']
                                  .value_counts())
winning_by_toss = pd.DataFrame({
    'Team': toss_winner_match_winner_count.index,
    'Wins': toss_winner_match_winner_count.values
})
winning_by_toss.plot(kind='bar', x='Team', y='Wins', title='Winning Analysis according to Toss')
```

```python
# Most wins by a team
most_wins_count = data['winner'].value_counts()
most_wins = most_wins_count.index[0]

print("Team with most wins:", most_wins)
```

```python
# Most loss by a team
most_loss_count = data['winner'].value_counts()
most_loss = most_loss_count.index[-1]

print("Team with most loss:", most_loss)
```

```python
# Win by highest margin (Runs)
highest_margin = data.loc[data['result_margin'].idxmax()]
print("Win by highest margin (Runs): ")
print(highest_margin[['winner', 'result_margin']])
```

```python
# Win by lowest margin (Runs)
lowest_margin = data.loc[data['result_margin'].idxmin()]
print("Win by lowest margin (Runs): ")
print(lowest_margin[['winner', 'result_margin']])
```

```python
# Visualize winnings of every team in bar graph
team_wins = data.groupby('winner')['id'].count().sort_values(ascending=False)
plt.bar(team_wins.index, team_wins.values)
plt.xticks(rotation=90)
plt.title("Winnings of every team")
plt.xlabel("Teams")
plt.ylabel("Number of wins")
plt.show()
```

```python
# Visualize the performance (batting or bowling) of given player in bar graph
player_name = 'V Kohli'
player_data = data[(data['player_of_match']==player_name)]
performance = player_data.groupby('winner')['id'].count()
plt.bar(performance.index, performance.values)
plt.xticks(rotation=90)
plt.title("Performance of " + player_name)
plt.xlabel("Teams")
plt.ylabel("Number of matches")
plt.show()
```
